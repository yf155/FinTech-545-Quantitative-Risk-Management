```python
from riskmgmt import *
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import data
data = pd.read_csv("problem1.csv")
x = data['x']

# Fit with normal distribution and generalized T distribution
norm = fit_norm(x)
gen_t = fit_general_t(x)
```

```python
# Calculate the VaR and ES for both fitted distributions
print("VaR for normal distribution:",VaR(norm))
print("ES for normal distribution:",ES(norm))
print("VaR for generalized T distribution:",VaR(gen_t))
print("ES for generalized T distribution:",ES(gen_t))
```

```python
#Overlay the graphs the distribution PDFs, VaR, and ES values
plt.hist(x, bins=100, density=True, alpha=0.5, label='Original data')
sns.kdeplot(norm,color="Green",label="Normal")
sns.kdeplot(gen_t,color="Orange",label="T")
plt.axvline(-VaR(norm), color="Red",label="Norm VaR")
plt.axvline(-VaR(gen_t), color="blue",label="T VaR",linestyle="--")
plt.axvline(-ES(norm), color="Red",label="Norm ES")
plt.axvline(-ES(gen_t), color="blue",label="T ES",linestyle="--")
plt.legend()
```
