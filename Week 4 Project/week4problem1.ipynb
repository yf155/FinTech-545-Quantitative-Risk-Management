```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Set Parameters
sigma = 0.1
p_0 = 100
n = 10000
```

```python
r_t = np.random.normal(0, sigma, size = n)
p_t = np.zeros(n)
```

```python
# Simulate Classic Brownian Motion
for i in range(n):
    p_t[i] = p_0 + r_t[i]
    
print("Expected mean of Classical Brownian Motion:",p_0)
print("Expected std of Classical Brownian Motion:",sigma)
print("mean of Classical Brownian Motion:",p_t.mean())
print("std of Classical Brownian Motion:",p_t.std())

plt.hist(p_t, bins=50, density=True)
plt.title("Classic Brownian Motion")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Simulate Arithmetic Return System
for i in range(n):
    p_t[i] = p_0 * (1 + r_t[i])
    
print("Expected mean of Arithmetic Return System:",p_0)
print("Expected std of Arithmetic Return System:",p_0 * sigma)
print("mean of Arithmetic Return System:",p_t.mean())
print("std of Arithmetic Return System:",p_t.std())

plt.hist(p_t, bins=50, density=True)
plt.title("Arithmetic Return System")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Simulate Geometric Brownian Motion
for i in range(n):
    p_t[i] = p_0 * math.exp(r_t[i])
    
print("Expected mean of Geometric Brownian Motion:",p_0 * math.exp((sigma**2)/2))
print("Expected std of Geometric Brownian Motion:",p_0 * math.sqrt((math.exp(sigma**2)-1)*math.exp(sigma**2)))
print("mean of Geometric Brownian Motion:",p_t.mean())
print("std of Geometric Brownian Motion:",p_t.std())

plt.hist(p_t, bins=50, density=True)
plt.title("Geometric Brownian Motion")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```
